```python
# найти коэффициенты регрессии.
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
# Разделите данные на признаки и целевую переменную
X = df.drop(['index', 'target'], axis=1)
y = df['target']

# Обучите модель линейной регрессии
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
# Функция для линейной регрессии
def linear_regression(X, y):
    X = np.array(X).tolist()  # Преобразуем X в формат списка
    y = np.array(y).tolist()  # Преобразуем y в формат списка
    X_transpose = np.transpose(X)  # Транспонируем матрицу X
    X_transpose_X_inv = np.linalg.inv(np.dot(X_transpose, X))  # Находим обратную матрицу (X^T * X)^(-1)
    X_transpose_y = np.dot(X_transpose, y)  # Умножаем транспонированную матрицу X на y
    coefficients = np.dot(X_transpose_X_inv, X_transpose_y)  # Вычисляем коэффициенты модели
    return coefficients
```

```python
coefficients = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

print("Коэффициенты регрессии:", coefficients)
```

```python
def predict(coefficients, X):
    predictions = []
    X = np.array(X).tolist()
    for i in range(len(X)):
        prediction = sum(coefficients[j] * X[i][j] for j in range(len(X[0])))
        predictions.append(prediction)
    return predictions

# Получение прогнозов
predictions = predict(coefficients, X_test)
predictions
```

```python
r2 = r2_score(y_test, predictions) 
r2
```
